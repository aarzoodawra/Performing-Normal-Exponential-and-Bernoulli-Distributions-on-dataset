```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import skew
import scipy.stats as stats
# Plot normal distribution areas
data = norm.rvs(10.0, 2 , size=200)
mn, std = norm.fit(data)
plt.hist(data, bins=30, normed=True, alpha=0.4, color='red')
min, max = plt.xlim()
x = np.linspace(min, max, 200)
p = norm.pdf(x, mn, std)
plt.plot(x, p, 'k', linewidth=2)
title = "Fit results: mn = %.2f,  std = %.2f" % (mn, std)
plt.title(title)
plt.show()
skew(data)
```

```python
# The graph shows the skewness of the normal distribution is right skewed.
```

```python

# exponential Fit
from scipy.stats import expon
from scipy.stats import skew
data= np.random.exponential(scale=0.2, size= 200)
loc, scale = expon.fit(data)
plt.hist(data, bins=30, normed=True, alpha=0.4, color='orange')
min, max = plt.xlim()
x = np.linspace(min,max,200)
p= expon.pdf(x, loc, scale)
plt.plot(x,p,'k',linewidth= 2)
plt.title(title)
plt.show()
skew(data)
```

```python
# The graph shows that curve is right skewed and higher than normal distribution.
```

```python
# Bernoulli fit
from scipy.stats import bernoulli
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
p = 0.3
mean, var, skew, kurt = bernoulli.stats(p, moments='mvsk')
x = np.arange(bernoulli.ppf(0.01, p),bernoulli.ppf(0.99, p))
ax.plot(x, bernoulli.pmf(x, p), 'bo', ms=8, label='bernoulli pmf')
ax.vlines(x, 0, bernoulli.pmf(x, p), colors='maroon', lw=30, alpha=0.5)
r = bernoulli.rvs(p, size=200)
plt.show()
```
